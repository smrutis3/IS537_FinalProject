```python
import pandas as pd
import numpy as np

# Load datasets
conc = pd.read_csv('annual_conc_by_monitor_2025.csv')
sites = pd.read_csv('aqs_sites.csv')

# Preview
print(conc.shape)
print(sites.shape)

conc.head()
```

```python
#data profiling
# Basic info
conc.info()
sites.info()

# Missing values
conc.isnull().sum().sort_values(ascending=False).head(20)
sites.isnull().sum().sort_values(ascending=False).head(20)

# Summary stats
conc.describe()
```

```python
#completeness check
# % missing values
missing_conc = conc.isnull().mean() * 100
missing_sites = sites.isnull().mean() * 100

missing_conc.sort_values(ascending=False).head(10)
missing_sites.sort_values(ascending=False).head(10)
```

```python
#checking duplicates
# Check duplicates
conc_duplicates = conc.duplicated().sum()
sites_duplicates = sites.duplicated().sum()

print("Conc duplicates:", conc_duplicates)
print("Sites duplicates:", sites_duplicates)
```

```python
#validity check
# Check extreme values (example column)
conc['Arithmetic Mean'].describe()

# Find suspicious values
conc[conc['Arithmetic Mean'] < 0]
conc[conc['Arithmetic Mean'] > 500]  # threshold depends on pollutant
```

```python
print(conc[['State Code', 'County Code', 'Site Num']].dtypes)
print(sites[['State Code', 'County Code', 'Site Num']].dtypes)
```

```python
# Convert keys to string in BOTH datasets
for col in ['State Code', 'County Code', 'Site Num']:
    conc[col] = conc[col].astype(str)
    sites[col] = sites[col].astype(str)
```

```python
print(conc.columns.tolist())
print(sites.columns.tolist())
```

```python
# Rename in sites dataset
sites = sites.rename(columns={'Site Number': 'Site Num'})
```

```python
#consistency check
# Create join keys
keys = ['State Code', 'County Code', 'Site Num']

merged = pd.merge(conc, sites, on=keys, how='left', indicator=True)

merged['_merge'].value_counts()
```

```python
merged['_merge'].value_counts(normalize=True)
```

```python
unmatched = merged[merged['_merge'] == 'left_only']

unmatched[['State Code', 'County Code', 'Site Num']].head()
```

```python
# Which states have most missing metadata?
unmatched['State Code'].value_counts().head(10)
```
